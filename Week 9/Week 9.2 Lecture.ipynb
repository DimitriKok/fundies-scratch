## Extracting Columns & Statistics

- Extract the duration column from your workouts DataFrame and store it in a variable called durations.

- Work with the durations Series to find: .mean(), .sum(), .max(), .min().

- Calculate the range (difference between max and min) of workout durations.

- For the photos dataset, extract a numeric column and calculate its median using .median().

- Get the value counts of the Subject from the photos DataFrame

- Calculate its median number of photos per subject using .median()

```python
durations = df["duration"]

# Prints the average
print(durations.mean())

# Prints the sun
print(durations.sum())

# Prints the maximum number
print(durations.max())

# Prints the minimum number
print(durations.min())

# Calculating the range
print(durations.max() - durations.min())

# Finds the median
print(df["duration"].median)
photos["Date"] = pd.to_datetime(photos["Date"])

print(photos["Date"].median())

# Gets the value counts of the subject from the photos dataframe
print(photos["Subject"].count())

# Calculating the median number of photos per subject using .median
```

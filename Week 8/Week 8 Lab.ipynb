Write a Python function `letter_grade` that assigns letter grades based on numeric scores:

- 90-100: `"A"`
- 80-89: `"B"`
- 70-79: `"C"`
- 60-69: `"D"`
- Below 60: `"F"`

````python
%%ipytest -qq

# Write code here
def letter_grade(score: float):
    if (score >= 90):
        return "A"
    elif (score >= 80):
        return "B"
    elif (score >= 70):
        return "C"
    elif (score >= 60):
        return "D"
    else:
        return "F"



def test_letter_grade():
    ## assert statments go here
    assert letter_grade(1) == "F"
    assert letter_grade(61) == "D"
    assert letter_grade(73) == "C"
    assert letter_grade(84) == "B"
    pass
````

## Problem 2
Design a function `c_to_f` that transforms a list of temperatures from Celsius to Fahrenheit, using the formula 

`F = C * 9/5 + 32`

Hint: You will need the `list` and `map` functions with a `lambda`

````python
# Write your code here
celcius = [41, 30, 28]

def c_to_f(l : list) -> list:
    return list(map(lambda n: n * 9/5 + 32, l))

c_to_f(celcius)
````

Design a function `long_words_upper that`, given a list of words (Strings), finds all words longer than `4` characters and convert them to uppercase. (Hint: map and filter may be helpful here!)

````python
"hello world!".upper()
words = ["absc", "sefsf", "cowss"]

def long_words_upper(l: list) -> list:
    longWordsOnly = list(filter(lambda wd: (len(wd) > 4), l))
    return list(map(lambda wd: wd.upper(), longWordsOnly))

long_words_upper(words)
````

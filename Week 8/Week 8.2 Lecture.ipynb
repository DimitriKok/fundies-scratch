````python
def student_grades(mark: float) -> str:
    """
    This function takes student marks and provide a grade!
    mark type: float; student marks
    return: string
    """
    if mark < 50:
        return "Fail"
    elif mark < 70:
        return "Pass"
    else:
        return "Distinction"
````
